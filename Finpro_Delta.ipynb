# **Background**

## **Business Understanding**

**Context :**

Dalam sektor perbankan, peran deposit sangat krusial. Deposit merupakan dana yang disimpan oleh nasabah di bank selama jangka waktu tertentu, dengan imbal hasil (return) karena telah meminjamkan uangnya. Bank memerlukan deposit sebagai sumber dana untuk memberikan pinjaman kepada nasabah lain dan menginvestasikan dana, yang pada akhirnya menghasilkan return yang lebih besar. Selain itu, deposit juga mendukung operasional harian bank dan menjaga likuiditas serta stabilitas keuangan. Melalui program pemasaran yang efektif, bank dapat mendorong partisipasi aktif nasabah dalam menyimpan dana melalui berbagai produk deposit yang ditawarkan. Tujuan utama program pemasaran adalah meningkatkan jumlah deposit yang masuk ke bank, yang pada gilirannya memberikan manfaat jangka panjang untuk pertumbuhan dan stabilitas keuangan bank.[sumber:financestrategist.com](https://www.financestrategists.com/banking/bank-account/bank-account-deposits/?ssp=1&darkschemeovr=1&setlang=id&cc=ID&safesearch=moderate)

**Problem Statement :**

Karena jumlah nasabah yang sangat besar, kampanye pemasaran akan memerlukan waktu dan biaya yang signifikan. Tanpa melakukan filter terhadap nasabah yang akan dihubungi, risiko biaya kampanye melebihi keuntungan dari deposito yang diperoleh dari nasabah yang setuju melakukan term deposit. Oleh karena itu, diperlukan efisiensi dalam kampanye agar bank dapat memperoleh deposit sebanyak mungkin dengan biaya dan waktu yang minimal.

**Goals :**

Sebagai Data Scientist, kami akan mencoba untuk membantu perusahaan menekan biaya kampanye dengan cara memahami karakteristik dari nasabah yang berpotensi untuk melakukan term deposit agar kampanye/marketing dapat lebih terarah, adapun beberapa pertanyaan yang harus dijawab agar dapat menemukan karakteristik tersebut adalah :
1. Karakteristik Nasabah Berdasarkan Informasi Personal:
* Umur: Nasabah yang cenderung melakukan term deposit mungkin memiliki usia yang lebih matang dan stabil secara finansial.
* Status Pernikahan: Pernikahan dapat mempengaruhi keputusan nasabah untuk berinvestasi. Mungkin nasabah yang sudah menikah lebih cenderung melakukan term deposit.
* Pendidikan: Nasabah dengan pendidikan yang lebih tinggi mungkin lebih cermat dalam mengelola keuangan dan memilih instrumen investasi.
* Pekerjaan: Pekerjaan dan stabilitas pekerjaan juga memainkan peran. Nasabah dengan pekerjaan tetap mungkin lebih cenderung melakukan term deposit.
* Pinjaman dan Kredit: Nasabah yang memiliki pinjaman atau kredit mungkin lebih berhati-hati dalam mengelola keuangan dan memilih instrumen investasi.

2. Karakteristik Nasabah Berdasarkan Metode Komunikasi:
* Nasabah yang akhirnya menyetujui term deposit mungkin lebih responsif terhadap komunikasi melalui telepon atau email daripada metode lainnya.
* Durasi panggilan terakhir sebelum nasabah memutuskan dapat menjadi indikator efektivitas komunikasi.

3. Karakteristik Nasabah Berdasarkan Perlakuan Bank pada Kampanye:
* Nasabah yang menerima term deposit mungkin telah menerima lebih banyak kontak dari bank pada kampanye sebelumnya.
* Keputusan nasabah pada kampanye sebelumnya (misalnya, apakah mereka menyetujui produk atau tidak) dapat memberikan wawasan tentang preferensi mereka.

4. Situasi Ekonomi Portugal:
* Faktor ekonomi seperti inflasi (CPI), Consumer Confidence Index (CCI), suku bunga euribor, tingkat variasi penerimaan kerja, dan indikator jumlah tenaga kerja dapat mempengaruhi keputusan nasabah untuk melakukan deposit.

## **Metric Evaluation :**

Karena fokus utama kita adalah nasabah yang berminat melakukan term deposit, maka kita menetapkan target sebagai berikut:

* 0 : Tidak melakukan term deposit
* 1 : Melakukan term deposit


Selain itu, kita perlu memahami dua jenis kesalahan yang mungkin terjadi dalam prediksi:

1. Type 1 error (False Positive): Ini terjadi ketika nasabah sebenarnya tidak akan melakukan term deposit, tetapi diprediksi akan melakukannya. Konsekuensinya adalah potensi pengeluaran biaya yang sia-sia karena Bank akan menghubungi nasabah tersebut padahal sebenarnya tidak ada minat untuk melakukan deposit.
2. Type 2 error (False Negative): Ini terjadi ketika nasabah sebenarnya akan melakukan term deposit, tetapi diprediksi tidak akan melakukannya. Konsekuensinya adalah kehilangan nasabah yang sebenarnya akan melakukan term deposit, yang dapat menambah keuntungan Bank, karena diputuskan untuk tidak menghubungi mereka berdasarkan prediksi yang salah.

Untuk memberikan gambaran secara kuantitatif, kita akan mencoba menghitung dampak biaya pada error diatas berdasarkan beberapa asumsi dibawah :

* Biaya Panggilan Telepon: Menurut [liveagent.com](https://www.liveagent.com/customer-support-glossary/cost-per-call/), biaya rata-rata untuk setiap panggilan telepon berkisar antara 2.53 EURO hingga 5.15 EURO.
* Keuntungan Term Deposit: [Investopedia.com](https://www.investopedia.com/terms/t/timedeposit.asp), menyatakan bahwa term deposit sangat menguntungkan bagi bank. Setiap deposit dari nasabah dapat digunakan oleh bank untuk meminjamkan uang ke nasabah lain atau menginvestasikannya pada sekuritas lain yang berpotensi memberikan keuntungan lebih tinggi daripada yang dibayarkan oleh nasabah.
* Frekuensi Telepon: Berdasarkan [blog.hubspot.com](https://blog.hubspot.com/sales/best-times-to-connect-with-leads-infographic), setidaknya kita perlu menelfon konsumen sebanyak 6 kali agar mereka tertarik pada produk yang kita tawarkan.
* Suku Bunga Deposit dan Lending di Portugal: Pada tahun 2014, suku bunga deposit di Portugal adalah 0.2% berdasarkan [tradingeconomics.com](https://tradingeconomics.com/portugal/interest-rate), sementara suku bunga lending berkisar sekitar 5.5% juga berdasarkan [tradingeconomics.com](https://tradingeconomics.com/portugal/bank-lending-rate).
* Deposit Berjangka: Menurut [withportugal.com](https://withportugal.com/en/immigration/depositos-portugal), deposito berjangka di Portugal diperbolehkan dengan rentang nominal antara 100 EURO hingga 250,000 EURO. Mengingat mayoritas pekerjaan dalam dataset adalah admin, dengan rata-rata gaji pertahun sekitar 18,780 EURO berdasarkan informasi dari [paylab.com](https://www.paylab.com/pt/salaryinfo/administration#:~:text=The%20gross%20salary%20range%20fordrastically%20among%20different%20job%20positions.), dan kebutuhan hidup pada tahun 2014 sekitar 12,752 EURO [numbeo.com](https://www.numbeo.com/cost-of-living/historical-data?itemId=105&city_id=6029&name_city_id=Lisbon%2C+Portugal), kita dapat mengasumsikan bahwa deposit yang dilakukan oleh satu nasabah admin adalah sekitar 60% dari sisa uangnya, yaitu 0.6 * 6028 EURO = 3617 EURO.


Berdasarkan asumsi di atas maka kita dapat coba kuantifikasi konsekuensi dari setiap type error  sebagai berikut :
- Type 1 Error (False Positive) Pengeluaran biaya --> Untuk setiap panggilan, kita asumsikan biayanya adalah 3 EURO dan kita harus menelfon sebanyak 6 x, jadi biaya yang dikeluarkan adalah 18 EURO per nasabah per tahun

- Type 2 Error (False Negative) Kerugian akibat kehilangan nasabah potensial --> Untuk setiap pelanggan yang melakukan term deposit sebesar 3,756 EURO dalam jangka waktu 1 tahun, maka bank akan mendapatkan keuntungan sebesar : 
    - Keuntungan = (Nominal Deposit * (Bunga pinjaman - Bunga Deposit))
    - Keuntungan = (3617 x (5.5%-0.2%)) = 192 EURO per tahun

Berdasarkan informasi ini, kita akan membuat suatu business metric yang merupakan peningkatan keuntungan jika kita memakai machine learning yang akan kita namakan sebagai Business Metric

Kesimpulan Analisis Umur dan Term Deposit:

2. Oleh karena itu, kita akan menggunakan uji hipotesis Mann-Whitney untuk menguji apakah perbedaan median ini signifikan secara statistik.

Hipotesis :

- Ho : Distribusi Data umur 'Yes' = Distribusi Data umur 'No'
- Ha : Distribusi Data umur 'Yes' != Distribusi Data umur 'No'

```python
from scipy.stats import mannwhitneyu

stat, pvalue = mannwhitneyu(
    df[df['y'] == 'yes']['age'],
    df[df['y'] == 'no']['age'],
    alternative = 'two-sided'
)
pvalue
```

```python
# conclusion :
if pvalue > 0.05 :
    print(f'''
          pvalue = {pvalue:.5f}. pvalue > 0.05. Gagal menolak Ho.
          Berdasarkan bukti yang ada, kita tidak memiliki cukup alasan untuk menyatakan bahwa distribusi usia orang yang menyetujui term 
          deposit berbeda dengan distribusi usia orang yang menolak term deposit.
          (Distribusi datanya berbeda signifikan)
          ''')
else :
    print(f'''
          pvalue = {pvalue:.5f}. pvalue <= 0.05. Berhasil menolak Ho.
          Berdasarkan bukti yang ada, kita dapat menyimpulkan bahwa distribusi usia orang yang menyetujui term deposit berbeda secara 
          signifikan dengan distribusi usia orang yang menolak term deposit
          (Distribusi datanya berbeda signifikan)
          ''')
```

Kesimpulan Analisis Pinjaman dan Term Deposit:
1. Berdasarkan visualisasi conversion rate, tidak terlihat perbedaan yang signifikan antara nasabah yang mengambil pinjaman dan yang tidak.
2. Namun, kita dapat mengamati bahwa orang-orang yang tidak mengambil pinjaman cenderung memiliki conversion rate yang lebih tinggi.
3. Untuk menguji hubungan ini secara lebih pasti, akan dilakukan uji hipotesis chi-square.

```python
df_loan = df[['loan','y']]
df_loan['yes'] = df_loan['y'].apply(lambda x : 1 if x == 'yes' else 0)
df_loan['no'] = df_loan['y'].apply(lambda x : 1 if x == 'no' else 0)
df_loan = df_loan.drop(columns = 'y')
df_loan_chi = df_loan.groupby('loan').sum()
df_loan_chi
```

Hipotesis :

* Ho: Tidak ada keterkaitan antara pinjaman yang diambil oleh nasabah dan kecenderungannya untuk melakukan term deposit.
* Ha: Terdapat keterkaitan antara pinjaman yang diambil oleh nasabah dan kecenderungannya untuk melakukan term deposit.

```python
from scipy.stats import chi2_contingency
stat, pvalue, dof, expected = chi2_contingency(df_loan_chi)

if pvalue > 0.05 :
    print(f'''
          pvalue = {pvalue:.5f}. pvalue > 0.05. Gagal menolak Ho.
          TIDAK cukup bukti untuk menyatakan antara Pinjaman dan Target ada keterkaitan
          (proporsinya tidak berbeda signifikan)
          ''')
else :
    print(f'''
          pvalue = {pvalue:.5f}. pvalue <= 0.05. Berhasil menolak Ho.
          Ada cukup bukti untuk menyatakan antara Pinjaman dan Target ada keterkaitan
          (proporsinya berbeda signifikan)
          ''')
```

Kesimpulan :
1. Hasil analisis menunjukkan bahwa tidak terdapat hubungan yang signifikan antara pinjaman yang diambil oleh seseorang dan kecenderungannya untuk melakukan deposit. 
2. Nilai p-value yang besar menunjukkan bahwa kedua kolom tersebut tidak memiliki keterkaitan yang relevan. Meskipun secara domain knowledge, kita mungkin mengharapkan adanya pengaruh, namun temuan ini menunjukkan sebaliknya.

Hipotesis :

- Ho : Hari kontak terakhir dan kecenderungan nasabah untuk melakukan term deposit tidak ada keterkaitan
- Ha : Hari kontak terakhir dan kecenderungan nasabah untuk melakukan term deposit ada keterkaitan
